**How can you produce a list of facilities that charge a fee to members, and that fee is less than 1/50th of the monthly maintenance cost? Return the facid, facility name, member cost, and monthly maintenance of the facilities in question.**

**¿Cómo se puede producir una lista de instalaciones que cobran una tarifa a los miembros, y esa tarifa es inferior a 1/50 del costo de mantenimiento mensual? Devuelva el facid, el nombre de la instalación, el costo del miembro y el mantenimiento mensual de las instalaciones en cuestión.**

```python
pd.read_sql_query(
    '''
    SELECT
        facid,
        name,
        membercost,
        monthlymaintenance
    FROM
        cd.facilities
    WHERE
        membercost > 0 AND (membercost < monthlymaintenance / 50.0)
    '''
,engine).head()
```

**7. How can you produce a list of members who joined after the start of September 2012? Return the memid, surname, firstname, and joindate of the members in question**

**¿Cómo se puede producir una lista de los miembros que se unieron después del comienzo de septiembre de 2012? Devuelva el nombre, apellido, nombre y fecha de ingreso de los miembros en cuestión**

**8. How can you produce an ordered list of the first 10 surnames in the members table? The list must not contain duplicates.**

**¿Cómo se puede producir una lista ordenada de los primeros 10 apellidos en la tabla de miembros? La lista no debe contener duplicados.**


```python
pd.read_sql_query(
    '''
    SELECT
        DISTINCT(surname)
    FROM
        cd.members
    ORDER BY
        surname
    LIMIT 10
    '''
,engine)#.head()
```

**11. Produce a list of the total number of slots booked per facility in the month of September 2012. Produce an output table consisting of facility id and slots, sorted by the number of slots.**

**Genere una lista del número total de espacios reservados por instalación en el mes de septiembre de 2012. Genere una tabla de salida que consista en la identificación de la instalación y los espacios, ordenados por el número de espacios.**

```python
pd.read_sql_query(
    '''
    SELECT
        facid,
        SUM(slots) AS "Total Espacios reservados"
    FROM
        cd.bookings
    WHERE
        starttime >= '2012-09-01' AND starttime < '2012-10-01'
    GROUP BY
        facid 
    ORDER BY
        SUM(slots);
    '''
,engine)
```

**12. Produce a list of facilities with more than 1000 slots booked. Produce an output table consisting of facility id and total slots, sorted by facility id.**

**Produzca una lista de instalaciones con más de 1000 espacios reservados. Genere una tabla de salida que consista en la identificación de la instalación y el total de espacios, ordenados por identificación de la instalación.**

```python
pd.read_sql_query(
    '''
    SELECT
        facid,
        SUM(slots) AS total_slots
    FROM
        cd.bookings
    GROUP BY
        facid
    HAVING
        SUM(slots) > 1000
    ORDER BY
        facid;
    '''
,engine)
```

**13. How can you produce a list of the start times for bookings for tennis courts, for the date '2012-09-21'? Return a list of start time and facility name pairings, ordered by the time.**

**¿Cómo se puede generar una lista de las horas de inicio de las reservas para las canchas de tenis, para la fecha '2012-09-21'? Devuelve una lista de emparejamientos de hora de inicio y nombre de instalación, ordenados por hora.**

```python
pd.read_sql_query(
    '''
    SELECT
        cd.bookings.starttime AS start,
        cd.facilities.name AS name
    FROM
        cd.facilities
    INNER JOIN
        cd.bookings
    ON 
        cd.facilities.facid = cd.bookings.facid
    WHERE
        cd.facilities.facid IN (0,1)
        AND cd.bookings.starttime >= '2012-09-21'
        AND cd.bookings.starttime < '2012-09-22'
    ORDER BY
        cd.bookings.starttime;
    '''
,engine)
```

**14. How can you produce a list of the start times for bookings by members named 'David Farrell'?**

**¿Cómo se puede producir una lista de las horas de inicio de las reservas de los miembros llamados 'David Farrell'?**

```python
pd.read_sql_query(
    '''
    SELECT
        cd.bookings.starttime
    FROM
        cd.bookings
    INNER JOIN
        cd.members
    ON
        cd.members.memid = cd.bookings.memid
    WHERE
        cd.members.firstname = 'David'
    AND
        cd.members.surname = 'Farrell';
    '''
,engine)
```
