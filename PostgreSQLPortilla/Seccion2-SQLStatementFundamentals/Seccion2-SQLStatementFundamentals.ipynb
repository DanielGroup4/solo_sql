### ORDER BY

**"ORDER BY" towards the end of a query, since we want to do any selection and filtering first, before finally sorting.**

**"ORDER BYY" hacia el final de una query queremos hacer cualquier selección y filtrado primero, antes de ordenar finalmente.**

**Use ASC to sort in ascending order**

**Use DESC to sort in descending order**

**If you leave it blanck, ORDER BY uses ASC by default**

```python
pd.read_sql_query(
    '''
    SELECT
        *
    FROM
        customer
    ORDER BY
        first_name DESC;
    '''
,engine).head(3)
```

```python
pd.read_sql_query(
    '''
    SELECT
        store_id, 
        first_name,
        last_name 
    FROM
        customer
    ORDER BY
        store_id, 
        first_name ASC;
    '''
,engine).head(5)
```

```python
# you can place the field of the table in ORDER BY
# and omit it in the SELECT
pd.read_sql_query(
    '''
    SELECT 
        first_name,
        last_name 
    FROM
        customer
    ORDER BY
        store_id, 
        first_name ASC;
    '''
,engine).head(5)
```

### LIMIT

**The LIMIT command allows us to limit the number of rows returned for a query**

**LIMIT also becomes useful in combination with ORDER BY**

### BETWEEN

**The BETWEEN operator can be used with dates.  Note that you need to format dates in the ISO 8601 standard format, which is YYYY-MM-DD**

**When using BETWWEEN operator with dates that also include timestamp information, pay carful attention to using BETWEEN verses <=, >= comparison operators, due to the fact that a datetime starts at 0:00**

**Al usar el operador BETWWEEN con fechas que también incluyen información de marca de tiempo, preste mucha atención al uso de operadores de comparación BETWEEN versos <=, >=, debido al hecho de que una fecha y hora comienza a las 0:00**

```python
pd.read_sql_query(
    '''
    SELECT
        COUNT(*)
    FROM
        payment
    WHERE
        amount BETWEEN 8 AND 9;
    '''
,engine).head(5)
```

```python
# if a wanted to know the opposte, essentially everything 
# that wasnt between eight and nine dollars
# si quisiera saber lo contrario, esencialmente todo eso no era
# entre ocho y nueve dolares
pd.read_sql_query(
    '''
    SELECT
        COUNT(*)
    FROM
        payment
    WHERE
        amount NOT BETWEEN 8 AND 9;
    '''
,engine)#.head(5)
```

```python
pd.read_sql_query(
    '''
    SELECT
        *
    FROM
        payment
    WHERE
        payment_date
    BETWEEN '2007-02-01' AND '2007-02-15';
    '''
,engine).head(5)
```

### IN

**In certain cases you want to check for multiple possible value opotions, for example, if a user´s name shows uo IN a list of known names.**

**We can use the IN operator to create a condition that checks to see if a value in included in a list of multiple options**

**En ciertos casos, desea verificar múltiples opciones de valor posibles, por ejemplo, si el nombre de un usuario se muestra IN una lista de nombres conocidos.**

**Podemos usar el operador IN para crear una condición que verifique si un valor está incluido en una lista de múltiples opciones**

```python
pd.read_sql_query(
    '''
    SELECT
        *
    FROM
        payment
    WHERE
        amount IN (0.99, 1.98, 1.99);
    '''
,engine).head(5)
```

```python
pd.read_sql_query(
    '''
    SELECT
        COUNT(*)
    FROM
        payment
    WHERE
        amount IN (0.99, 1.98, 1.99);
    '''
,engine).head(5)
```

```python
pd.read_sql_query(
    '''
    SELECT
        COUNT(*)
    FROM
        payment
    WHERE
        amount NOT IN (0.99, 1.98, 1.99);
    '''
,engine).head(5)
```

```python
pd.read_sql_query(
    '''
    SELECT
        *
    FROM
        customer
    WHERE
        first_name IN ('John', 'Jake', 'Julie');
    '''
,engine).head(5)
```

### LIKE AND ILIKE

**The LIKE operator allows us to perform pattern matching against string data with the use of wildcards characters:**

**El operador LIKE nos permite realizar coincidencias de patrones con datos de cadena con el uso de caracteres comodín:**

* Percent % - Matches any sequence of characters *

* Underscore _ - Matches any single character *

**LIKE is case-sensitive, we can use ILIKE which is case-insensitive**

**Using the underscore " _ " allows us to replace just a single character**



```python
# in this case you need to type 2 times  %
pd.read_sql_query(
    '''
    SELECT
        *
    FROM
        customer
    WHERE
        first_name LIKE 'J%%';
    '''
,engine)
```

```python
pd.read_sql_query(
    '''
    SELECT
        *
    FROM
        customer
    WHERE
        first_name LIKE 'J%%' AND last_name LIKE 'S%%';
    '''
,engine)
```

```python
# get back nothing because LIKE is case sensitive
pd.read_sql_query(
    '''
    SELECT
        *
    FROM
        customer
    WHERE
        first_name LIKE 'j%%' AND last_name LIKE 's%%';
    '''
,engine)
```

```python
# ILIKE is not case sensitive
pd.read_sql_query(
    '''
    SELECT
        *
    FROM
        customer
    WHERE
        first_name ILIKE 'j%%' AND last_name ILIKE 's%%';
    '''
,engine)
```

```python
pd.read_sql_query(
    '''
    SELECT
        *
    FROM
        customer
    WHERE
        first_name LIKE '%%her%%';
    '''
,engine)
```

```python
pd.read_sql_query(
    '''
    SELECT
        *
    FROM
        customer
    WHERE
        first_name LIKE '_her%%';
    '''
,engine)
```

```python
pd.read_sql_query(
    '''
    SELECT
        *
    FROM
        customer
    WHERE
        first_name NOT LIKE '_her%%';
    '''
,engine)
```

```python
pd.read_sql_query(
    '''
    SELECT
        *
    FROM
        customer
    WHERE
        first_name LIKE 'A%%'
    ORDER BY
        last_name
    LIMIT 7;
    '''
,engine)
```
